```python
columns = df.columns

# Crear una función para hacer únicos los valores en el primer nivel
def make_unique(index):
    seen = {}
    new_index = []
    for item in index:
        if item in seen:
            seen[item] += 1
            new_index.append(f"{item}_{seen[item]}")  # Añadir un sufijo para hacerlo único
        else:
            seen[item] = 0
            new_index.append(item)
    return new_index

# Modificar el índice para asegurarse de que los valores en el nivel 0 sean únicos
new_level_0 = make_unique(columns.get_level_values(0))

# Crear el nuevo MultiIndex
new_columns = pd.MultiIndex.from_arrays(
    [new_level_0, columns.get_level_values(1)], 
    names=columns.names
)

# Asignar el nuevo índice de columnas al DataFrame
df.columns = new_columns

# Verificar el cambio
print(df.columns)
```

```python
# Cambiar 'bodypart' por 'bodyparts' solo en el primer nivel del MultiIndex
df.columns = df.columns.set_levels(
    df.columns.get_level_values(0).map(lambda x: 'bodyparts' if x == 'bodypart' else x),
    level=0
)

# Verificar el cambio
print(df.columns)
```

```python
# Cambiar 'bodypart' por 'bodyparts' en el primer nivel del MultiIndex
df.columns = df.columns.set_levels(
    df.columns.get_level_values(0).map(lambda x: 'bodyparts' if x == 'bodypart' else x), 
    level=0
)

# Verificar el cambio
print(df.columns)
```

```python
df.columns = df.columns.set_levels(
    [name.replace('bodypart', 'bodyparts') for name in df.columns.levels[0]],
    level=0)
```

```python
print(df.columns.names)  # Nombres de los niveles del MultiIndex
print(df.columns.levels)  # Valores en cada nivel del MultiIndex
```

```python
df.columns.names = ['bodyparts' if name == 'bodypart' else name for name in df.columns.names]
```

```python
df.columns = df.columns.set_levels(
    ["nose" if "nose" in part else part for part in df.columns.get_level_values("bodyparts")], 
    level="bodyparts"
)
```
